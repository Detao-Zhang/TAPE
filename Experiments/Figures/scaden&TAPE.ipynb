```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
```

```python
# comparison on human-ROSMAP data
hROSMAP_s = np.load("100_human_ROSMAP_scaden.npy")
hROSMAP_ae = np.load("100_human_ROSMAP_AE.npy")
```

```python
sns.scatterplot(hROSMAP_s[:,0],hROSMAP_s[:,1])
sns.scatterplot(hROSMAP_ae[:,0],hROSMAP_ae[:,1])
plt.xlabel("L1 error rate")
plt.ylabel("CCC")
plt.ylim(0.6,1)
plt.xlim(0.1,0.)
plt.legend(["Scaden","AE"],loc='upper left')
plt.title("Deconvolution result of ROSMAP dataset")
```

```python
colors = sns.color_palette()
fig, axes = plt.subplots(1,5,figsize=(24,4))

sns.scatterplot(pbmc_sdy67_ae[:,0],pbmc_sdy67_ae[:,1],ax=axes[0],color=colors[0])
sns.scatterplot(pbmc_sdy67_s[:,0],pbmc_sdy67_s[:,1],ax=axes[0],color=colors[1])
axes[0].set_xlim(0.2,0.05)
axes[0].set_ylabel('CCC')
x_major_locator=MultipleLocator(0.05)
axes[0].xaxis.set_major_locator(x_major_locator)

sns.scatterplot(pbmc_monaco_ae[:,0],pbmc_monaco_ae[:,1],ax=axes[1],color=colors[0])
sns.scatterplot(pbmc_monaco_s[:,0],pbmc_monaco_s[:,1],ax=axes[1],color=colors[1])
axes[1].set_xlim(0.1,0.04)
x_major_locator=MultipleLocator(0.02)
axes[1].xaxis.set_major_locator(x_major_locator)

sns.scatterplot(pbmc_microarray_ae[:,0],pbmc_microarray_ae[:,1],ax=axes[2],color=colors[0])
sns.scatterplot(pbmc_microarray_s[:,0],pbmc_microarray_s[:,1],ax=axes[2],color=colors[1])
axes[2].set_xlim(0.15,0.06)
x_major_locator=MultipleLocator(0.03)
axes[2].xaxis.set_major_locator(x_major_locator)

sns.scatterplot(hROSMAP_ae[:,0],hROSMAP_ae[:,1],ax=axes[3],color=colors[0])
sns.scatterplot(hROSMAP_s[:,0],hROSMAP_s[:,1],ax=axes[3],color=colors[1])
axes[3].set_xlim(0.09,0.03)
x_major_locator=MultipleLocator(0.02)
axes[3].xaxis.set_major_locator(x_major_locator)

sns.scatterplot(mROSMAP_ae[:,0],mROSMAP_ae[:,1],ax=axes[4],color=colors[0])
sns.scatterplot(mROSMAP_s[:,0],mROSMAP_s[:,1],ax=axes[4],color=colors[1])
axes[4].set_xlim(0.1,0.04)
x_major_locator=MultipleLocator(0.02)
axes[4].xaxis.set_major_locator(x_major_locator)

for i in range(5):
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)
handles, labels = axes[0].get_legend_handles_labels()

fig.legend(["TAPE","Scaden"],loc='upper right')
fig.xlabel('Mean absolute error')
# plt.tight_layout()

fig.savefig('TAPE_Scaden.eps', format='eps', dpi=300)

plt.show()
```

```python
microarray_ae = np.sum((pbmc_microarray_ae - np.array([0,1]))**2,axis=1)
microarray_s = np.sum((pbmc_microarray_s - np.array([0,1]))**2,axis=1)
t,p_twoTail=stats.ttest_rel(microarray_ae,microarray_s)
print(p_twoTail/2)
```
