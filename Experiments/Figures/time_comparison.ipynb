```python
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
# mpl.rcParams['font.sans-serif'] = ['SimHei']

names = ['100', '200', '300', '400', '500','600','700','800']
x = range(len(names))
tape = [112.723551, 107.582993, 110.445538, 106.561099, 114.789673, 146.632159, 157.503918, 122.818625]
rnasieve = [120.199867, 158.519182, 198.17657, 231.025254, 275.395032, 315.850106, 348.131951, 398.682145]
dwls = [267.849, 530.958, 839.188, 1064.646, 1314.96, 1581.759, 1913.177, 2133.687]
scaden = [179.16208, 191.712773, 197.823405, 192.400318, 186.193067, 190.963967, 232.230936, 181.946578]
csx = [821.68, 1664.05, 2595.24]

#pl.xlim(-1, 11)  # 限定横轴的范围
plt.ylim(-1, 2000)  # 限定纵轴的范围
plt.plot(x, tape, marker='o', mec='y', mfc='w',label=u'TAPE')
plt.plot(x, rnasieve, marker='*', ms=10,label=u'RNASieve')
plt.plot(x, dwls, marker='s', ms=5,label=u'DWLS')
plt.plot(x, scaden, marker='d', ms=5,label=u'Scaden')
plt.plot(range(len(csx)), csx, marker='x', ms=5,label=u'CSx')


plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=0)
# plt.margins(0)
# plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"sample") #X轴标签
plt.ylabel("time(s)") #Y轴标签
plt.title("time_comparison") #标题
plt.savefig('time_comparison.eps')
plt.show()
```
