```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
names = ['100', '200', '300', '400', '500','600','700','800']
x = range(len(names))
tape = [112.723551, 107.582993, 110.445538, 106.561099, 114.789673, 146.632159, 157.503918, 122.818625]
rnasieve = [120.199867, 158.519182, 198.17657, 231.025254, 275.395032, 315.850106, 348.131951, 398.682145]
dwls = [267.849, 530.958, 839.188, 1064.646, 1314.96, 1581.759, 1913.177, 2133.687]
scaden = [340.16208, 352.71277, 358.823405, 353.4003, 347.1930, 351.963967, 393.230936, 342.946578]
csx = [821.68, 1664.05, 2595.24]
music = [14.735, 17.21279, 21.19709, 26.13245, 28.72774, 34.14805, 39.23214, 44.02474]
bisque = [12.39346, 12.42249, 12.8113, 13.4531, 13.8746, 14.67, 14.98834, 15.13714]
```

```python
colors = sns.color_palette()
colors = [colors[i] for i in [0,1,4,3,5,6,9]]
```

```python
fig, ax = plt.subplots(figsize=(6,5))

sns.lineplot(x, np.log(tape), marker=',', ax=ax, color=colors[0])
sns.lineplot(x, np.log(scaden), marker=',', ax=ax, color=colors[1])
sns.lineplot(x, np.log(rnasieve), marker=',', ax=ax, color=colors[2])
sns.lineplot(range(len(csx)), np.log(csx), marker=',', ax=ax, color=colors[3])
sns.lineplot(x, np.log(dwls), marker=',', ax=ax, color=colors[4])
sns.lineplot(x, np.log(music), marker=',', ax=ax, color=colors[5])
sns.lineplot(x, np.log(bisque), marker=',', ax=ax, color=colors[6])

plt.legend(["TAPE","Scaden","RNAsieve","CIBERSORTx","DWLS","MuSiC","Bisque"],loc='best')
plt.xticks(x, names, rotation=0)
plt.xlabel(u"sample number") #X轴标签
plt.ylabel("Log time") #Y轴标签
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
fig.savefig('time_comparison.eps', format='eps', dpi=300)
plt.show()
```
