```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pbmc = pd.read_csv('PBMCseries_real_result.csv',index_col=0)
ROSMAP = pd.read_csv('ROSMAP_result.csv',index_col=0)
real = pd.concat([pbmc,ROSMAP],axis=0)
```

```python
tape = real[real['Method']=='TAPE']
```

```python
colors = sns.color_palette()
```

```python
plt.figure()
sns.catplot(x="Method", y="CCC",
            col="Dataset",
            data=real, kind="box",
            order = ['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]],
            fliersize=0.1,linewidth=1,
            height=5, aspect=1)
plt.savefig('real_ccc.eps', format='eps', dpi=300)
```

```python
plt.figure()
sns.catplot(x="Method", y="L1error",
            col="Dataset",
            data=real, kind="box",
            order = ['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]],
            fliersize=0.1,linewidth=1,
            height=5, aspect=1.)
plt.savefig('real_l1.eps', format='eps', dpi=300)
```

```python
CD8 = real.loc[real['CellType']=='CD8Tcells']
CD4 = real.loc[real['CellType']=='CD4Tcells']
similar = pd.concat([CD4,CD8],axis=0)
```
