```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = ['100', '200', '300', '400', '500','600','700','800']
x = range(len(names))
tape = [112.723551, 107.582993, 110.445538, 106.561099, 114.789673, 146.632159, 157.503918, 122.818625]
rnasieve = [120.199867, 158.519182, 198.17657, 231.025254, 275.395032, 315.850106, 348.131951, 398.682145]
dwls = [267.849, 530.958, 839.188, 1064.646, 1314.96, 1581.759, 1913.177, 2133.687]
scaden = [179.16208, 191.712773, 197.823405, 192.400318, 186.193067, 190.963967, 232.230936, 181.946578]
csx = [821.68, 1664.05, 2595.24]
```

```python
colors = sns.color_palette()
colors = [colors[i] for i in [0,1,4,3,5]]
```

```python
fig, ax = plt.subplots(figsize=(6,4))

sns.lineplot(x, tape, marker='o', ax=ax, color=colors[0])
sns.lineplot(x, scaden, marker='o', ax=ax, color=colors[1])
sns.lineplot(x, rnasieve, marker='o', ax=ax, color=colors[2])
sns.lineplot(range(len(csx)), csx, marker='o', ax=ax, color=colors[3])
sns.lineplot(x, dwls, marker='o', ax=ax, color=colors[4])



plt.legend(["TAPE","Scaden","RNAsieve","CSx","DWLS"],loc='best')
plt.xticks(x, names, rotation=0)
plt.xlabel(u"sample number") #X轴标签
plt.ylabel("time(s)") #Y轴标签
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('time_comparison.eps', format='eps', dpi=300)
plt.show()
```
