```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# vis normal test
normal = pd.read_csv('pseudobulk_normal_result.csv',index_col=0)
# vis rare test
rare = pd.read_csv('pseudobulk_rare_result.csv',index_col=0)
rare = rare.loc[rare['abundant'] == 'rare']
# vis similar test
sim = pd.read_csv('pseudobulk_similar_result.csv',index_col=0)
sim = sim.loc[sim['CellType'] == 'Bcell_sum']
# vis lpb&ib proportions in common
bcell = pd.read_csv('pseudobulk_normal_result.csv',index_col=0)
bcell = pd.concat([bcell.loc[bcell['CellType']=='late pro-B cell'],bcell.loc[bcell['CellType']=='immature B cell']])
```

```python
print(normal[normal['Method'] == 'TAPE'].shape)
print(rare[rare['Method'] == 'TAPE'].shape)
print(bcell[bcell['Method'] == 'TAPE'].shape)
print(sim[sim['Method'] == 'TAPE'].shape)
```

```python
colors = sns.color_palette()
```

```python
# Overall
fig, axes = plt.subplots(2,4, sharey=False, figsize=(20, 10))
sns.boxplot(x="Method", y="CCC", data=normal, 
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[0,0],
            fliersize=0.1,linewidth=1)

axes[0,0].spines['top'].set_visible(False)
axes[0,0].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="L1error", data=normal,
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[1,0],
            fliersize=0.1,linewidth=1)

axes[1,0].spines['top'].set_visible(False)
axes[1,0].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="CCC", data=rare, 
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[0,1],
            fliersize=0.1,linewidth=1)

axes[0,1].spines['top'].set_visible(False)
axes[0,1].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="L1error", data=rare,
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[1,1],
            fliersize=0.1,linewidth=1)
axes[1,1].spines['top'].set_visible(False)
axes[1,1].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="CCC", data=bcell, 
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[0,2],
            fliersize=0.1,linewidth=1)

axes[0,2].spines['top'].set_visible(False)
axes[0,2].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="L1error", data=bcell,
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[1,2],
            fliersize=0.1,linewidth=1)
axes[1,2].spines['top'].set_visible(False)
axes[1,2].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="CCC", data=sim, 
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[0,3],
            fliersize=0.1,linewidth=1)

axes[0,3].spines['top'].set_visible(False)
axes[0,3].spines['right'].set_visible(False)

sns.boxplot(x="Method", y="L1error", data=sim,
            order=['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            palette = [colors[i] for i in [0,1,4,3,5,6,9]], ax=axes[1,3],
            fliersize=0.1,linewidth=1)
axes[1,3].spines['top'].set_visible(False)
axes[1,3].spines['right'].set_visible(False)

fig.savefig('psuedo_overall.eps', format='eps', dpi=300)
plt.show()
```

```python
plt.figure()
sns.catplot(x="Method", y="CCC",
            hue="DataType", col="Tissue",
            data=normal, kind="box",
            order = ['TAPE','Scaden','RNAsieve','CIBERSORTx','DWLS', 'MuSiC','Bisque'],
            fliersize=0.1,linewidth=1,
            height=5, aspect=1.3)
plt.savefig('psuedo_normal_ccc.eps', format='eps', dpi=300)
```
