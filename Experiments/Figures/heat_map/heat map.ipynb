```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
```

```python
ccc_real = pd.read_csv('/Users/chan/Desktop/submission_figdata/fig2/overallCCC.csv',index_col=0)
mae_real = pd.read_csv('/Users/chan/Desktop/submission_figdata/fig2/overallMAE.csv',index_col=0)
```

```python
fig, axes = plt.subplots(1,2,sharex=True,figsize=(10,5))

sns.heatmap(data=ccc_real, cmap='RdYlBu', square=True,
            annot=True, fmt=".2f", cbar=False, ax=axes[0])

sns.heatmap(data=mae_real, cmap='RdYlBu_r', square=True,
            annot=True, fmt=".2f", cbar=False, ax=axes[1])

# axes[0].xaxis.set_visible(False)
axes[1].yaxis.set_visible(False)

axes[1].set_xticklabels(labels=mae_real.columns,rotation=50)
axes[0].set_xticklabels(labels=mae_real.columns,rotation=50)
plt.tight_layout()
fig.savefig('overallperformance.eps', format='eps', dpi=300)
plt.show()
```
