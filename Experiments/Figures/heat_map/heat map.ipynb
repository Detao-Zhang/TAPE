```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
```

```python
ccc_real = pd.read_excel('ccc_real.xlsx',index_col=0)
ccc_pseudo = pd.read_excel('ccc_pseudo.xlsx',index_col=0)
mae_real = pd.read_excel('mae_real.xlsx',index_col=0)
mae_pseudo = pd.read_excel('mae_pseudo.xlsx',index_col=0)
```

```python
fig, axes = plt.subplots(2,2,sharex=True,figsize=(6,6))

sns.heatmap(data=ccc_pseudo, cmap='RdYlBu', square=True,
            annot=True, fmt=".2f", cbar=False, ax=axes[0,0])

sns.heatmap(data=ccc_real, cmap='RdYlBu', square=True,
            annot=True, fmt=".2f", cbar=False, ax=axes[1,0])

sns.heatmap(data=mae_pseudo, cmap='RdYlBu_r', square=True,
            annot=True, fmt=".2f", cbar=False, ax=axes[0,1])

sns.heatmap(data=mae_real, cmap='RdYlBu_r', square=True,
            annot=True, fmt=".2f", cbar=False, ax=axes[1,1])

axes[0,0].xaxis.set_visible(False)
axes[0,1].xaxis.set_visible(False)
axes[0,1].yaxis.set_visible(False)
axes[1,1].yaxis.set_visible(False)
axes[1,1].set_xticklabels(labels=mae_real.columns,rotation=50)
axes[1,0].set_xticklabels(labels=mae_real.columns,rotation=50)
plt.tight_layout()
fig.savefig('performance.eps', format='eps', dpi=300)
plt.show()
```
